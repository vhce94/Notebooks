```python
plt.figure(figsize = (15, 30))
for i, col in enumerate(df.columns, 1):
    plt.subplot(8, 2, i)
    sns.histplot(df[col], kde = True)
```

```python
ax = plt.figure(figsize=(18,6))
ax = plt.subplot(1,2,1)
ax = sns.countplot(x='LOCATION', data=df)
ax.bar_label(ax.containers[0])
plt.title("Ubicación del transformador", fontsize=20,color='Black',font='Times New Roman')
plt.legend(labels=['Rural','Urbana']);
plt.axis("off");

ax =plt.subplot(1,2,2)
ax=df['LOCATION'].value_counts().plot.pie(explode=[0.1, 0.1],autopct='%1.2f%%',shadow=True);
ax.set_title(label = "Ubicación del transformador", fontsize = 20,color='Black',font='Times New Roman');
plt.legend();
plt.axis("off");
```

```python
sns.displot(data=df, x="POWER", kde=True, bins = 100,color = "red", facecolor = "yellow",height = 5, aspect = 3.5);
```

```python
sns.displot(data=df, x="Average earth discharge density DDT [Rays/km^2-año]", kde=True, bins = 100,color = "red",
            facecolor = "yellow",height = 5, aspect = 3.5);
```

```python
ax = plt.figure(figsize=(15,7))
ax = plt.subplot(1,2,1)
ax = sns.countplot(x='Burned transformers 2019', data=df)
ax.bar_label(ax.containers[0])
plt.title("Estado del transformador", fontsize=1,color='Black',font='Times New Roman')
plt.legend(labels=['Transformador en buen estado']);
plt.axis("off");

ax =plt.subplot(1,2,2)
ax=df['Burned transformers 2019'].value_counts().plot.pie(explode=[0.1, 0.1],autopct='%1.2f%%',shadow=True);
ax.set_title(label = "Estado del transformador", fontsize = 20,color='Black',font='Times New Roman');
plt.legend();
plt.axis("off");
```

```python
ax = plt.figure(figsize=(18,6))
ax = sns.countplot(x="Type of clients", data=df)
ax.bar_label(ax.containers[0])
plt.title("Tipos de clientes", fontsize= 30,color='Black',font='Times New Roman');
```

```python
ax = plt.figure(figsize=(18,6))
ax = sns.countplot(x="Type of installation", data=df)
ax.bar_label(ax.containers[0])
plt.title("Tipos de clientes", fontsize= 30,color='Black',font='Times New Roman');
```
